```python
#import the libaries 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the datasets
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#remove/ delete the unwanted columns
df = dataset.drop(["Long","Lat","Province/State"], axis= 1, inplace = True)
```

```python
dataset.head()
```

```python
#aggregate the rows by country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualisation of data related to the country
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Qatar"].plot()
corona_dataset_aggregated.loc["Pakistan"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["India"][:50].plot()
```

```python
#calculate the 1st derivate of the curve
corona_dataset_aggregated.loc["India"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Pakistan"].diff().max()
```

```python
corona_dataset_aggregated.loc["Qatar"].diff().max()
```

```python
corona_dataset_aggregated
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rated"] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rated"])
```

```python
corona_data
```

```python
#import the 2nd dataset
happiness_report = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
#drop the useless columns
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols,axis= 1, inplace= True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region", inplace= True)
happiness_report.head()
```

```python
#join the dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report, how = "inner")
data
```
